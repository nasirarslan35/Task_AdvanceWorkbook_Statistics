```python
import numpy as np
import matplotlib.pyplot as plt
import math

lamda = 88
probabilities = []
x_values = []
for x in range(1, 100):
    prob = (lamda ** x * math.exp(-lamda)) / math.factorial(x)
    probabilities.append(prob)
    x_values.append(x)
print(probabilities)
# Median formula for Poisson distribution
median = lamda + (1/3) - (0.02/lamda)  

# Set the figure size
plt.figure(figsize=(14, 8))  # Example: 14 inches by 8 inches

plt.stem(x_values, probabilities)  # Create a stem plot
plt.title('Poisson Distribution PMF with $\lambda=88$')

# Find the first bar where probability is greater than 0.005 and the bar before that
first_found = False
for i, prob in enumerate(probabilities):
    if prob > 0.005 and not first_found:
        # Annotate this bar and the one before it (if it's not the very first bar)
        if i > 0:  # Check to avoid indexing error for the very first bar
            plt.text(x_values[i-1], probabilities[i-1], f'(X={x_values[i-1]}, P({x_values[i-1]})={(probabilities[i-1]*100):.3f}%)', 
                     ha='right', va='bottom')
        plt.text(x_values[i], prob, f'(X={x_values[i]}, P({x_values[i]})={(prob*100):.2f}%)', 
                 ha='right', va='bottom')
        first_found = True
        break  # Stop after annotating the required bars

# Display the median with a vertical line
plt.axvline(x=median, color='red', linestyle='--', label=f'Median: {median:.2f}')
plt.legend()

plt.xlabel('Number of Events (X)')
plt.ylabel('Probability')
plt.show()
```
